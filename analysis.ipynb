```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv("British Columbia_Full Data_data.csv", encoding="utf-16le", sep='\t')
df.head()
```

```python
# Display basic information about the dataset
print(f"Dataset shape: {df.shape}")
print(f"\nColumn names:")
print(df.columns.tolist())
```

```python
# Summary statistics
df.describe()
```

```python
  # Check missing values
  df.isnull().sum()
```

```python
  # Value counts for key columns
  df['Crash Severity'].value_counts()
  df['Day Of Week'].value_counts()
```

```python
  # Visualizations
  import matplotlib.pyplot as plt
```

```python
  # Crashes by year
  df['Date Of Loss Year'].value_counts().sort_index().plot(kind='bar', figsize=(10,6))
  plt.title('Crashes by Year')
  plt.xlabel('Year')
  plt.ylabel('Number of Crashes')
  plt.show()
```

```python
  # Crashes by day of week
  df['Day Of Week'].value_counts().plot(kind='bar', figsize=(10,6))
  plt.title('Crashes by Day of Week')
  plt.show()
```

```python
  # Crashes by severity
  df['Crash Severity'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8,8))
  plt.title('Crash Severity Distribution')
  plt.show()
```
